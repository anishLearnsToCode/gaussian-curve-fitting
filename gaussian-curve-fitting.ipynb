```python
from pydataset import data # to create dataset
import numpy as np #for defining mean and standard deviation
from matplotlib import pyplot as plt # for plotting graphs and visualisation
from scipy.stats import norm # for normal distribution
```

```python
%matplotlib inline 
```

```python
plt.hist(data('cancer')['age']) #selected age since age is normally distributed
plt.show()
```

```python
#sample standard deviation and mean
std= np.std(data('cancer')['age'], ddof=1)
mean= np.mean(data('cancer')['age'])
```

```python
#plotting
domain=np.linspace(np.min(data('cancer')['age']),np.max(data('cancer')['age']))
plt.plot(domain, norm.pdf(domain,mean,std),
         label='$\mathcal{N}$' + f'$( \mu \\approx {round(mean)} , \sigma \\approx {round(std)} )$') #latex label
plt.hist(data('cancer')['age'], edgecolor='black',alpha=.5,density=True)
plt.title("Gaussian Fit")
plt.xlabel("Value")
plt.ylabel("Density")
plt.legend()
plt.show()
```
